```python
SKU_List = sales1DF['SKU'].unique()
```

```python
outlet1 = []
for x in range(len(SKU_List)):
    tmp = pd.read_excel('sales_predict.xlsx',sheet_name='outlet1_{}'.format(SKU_List[x]))
    tmp['SKU'] = SKU_List[x]
    if x == 0:
        outlet1 = tmp
    else:
        outlet1 = outlet1.append(tmp)
```

```python
outlet2 = []
for x in range(len(SKU_List)):
    tmp = pd.read_excel('sales_predict.xlsx',sheet_name='outlet2_{}'.format(SKU_List[x]))
    tmp['SKU'] = SKU_List[x]
    if x == 0:
        outlet2 = tmp
    else:
        outlet2 = outlet2.append(tmp)
```

```python
outlet3 = []
for x in range(len(SKU_List)):
    tmp = pd.read_excel('sales_predict.xlsx',sheet_name='outlet3_{}'.format(SKU_List[x]))
    tmp['SKU'] = SKU_List[x]
    if x == 0:
        outlet3 = tmp
    else:
        outlet3 = outlet3.append(tmp)
```

```python
outlets = [outlet1, outlet2, outlet3]
```

```python
for df in outlets:
    df.sort_values(by='date',inplace=True)

outlets
```

```python
mush_combo = ['A001','A002','A003','A004']
veg_combo = ['A007','A008','A009','A010','A011']
sales_spec = [[2.0,1.0],[1.0,0.5],[0.085,0.043],[0.15,0.075],[1.0,0.5],[1.0,0.5],[0.125,0.065],[0.125,0.065],[0.15,0.075],[0.1,0.05],[0.2,0.1],[0.14]] 
```

```python
tempLists = [[],[],[]]
for j in range(len(outlets)):
    df = outlets[j]
    for i in range(len(df)):
        sku = df['SKU'].iloc[i] 
        if (sku == 'A013'):
            amt = df['sales'].iloc[i]
            dos = df['date'].iloc[i]
            
            for mush in mush_combo:
                index = int(mush[-2:]) - 1
                tempLists[j].append([sku,mush,dos,amt*sales_spec[index][1]])
                
        elif (sku=='A014'):
            amt = df['sales'].iloc[i]
            dos = df['date'].iloc[i]
            
            for veg in veg_combo:
                index = int(veg[-2:]) - 1
                tempLists[j].append([sku,veg,dos,amt*sales_spec[index][1]])
                
        elif (len(sku)>4 and sku[-2:] == '-2'):
            #A001-2 etc etc 
            amt = df['sales'].iloc[i]
            dos = df['date'].iloc[i]
            index = int(sku[2:4]) -1 
            tempLists[j].append([sku,sku[:4],dos,amt*sales_spec[index][1]])
            
        else:
            #A001-1 or A012-1
            amt = df['sales'].iloc[i]
            dos = df['date'].iloc[i]
            index = int(sku[2:4]) -1 
            tempLists[j].append([sku,sku[:4],dos,amt*sales_spec[index][0]])

tempLists
```

```python
header = pd.DataFrame(columns=['SKU','SPU','Date','Required Order'])
SPUsalesDFs = [header,header,header]

for i in range(len(tempLists)):
    #print(tempLists[i])
    tempDF = pd.DataFrame(tempLists[i],columns=('SKU','SPU','Date','Required Order'))            
    SPUsalesDFs[i] = SPUsalesDFs[i].append(tempDF)
SPUsalesDFs
```

```python
header = pd.DataFrame(columns=['SPU','Date','Required Order'])
spuDFs = [header,header,header]
for i in range(len(SPUsalesDFs)):
    spuDFs[i] = SPUsalesDFs[i].drop(columns=['SKU'],axis=1)
    spuDFs[i] = spuDFs[i].groupby(['Date','SPU'],as_index=False).agg({'Required Order': 'sum'})
    spuDFs[i] = spuDFs[i].sort_values(by=['SPU','Date'])

spuDFs
```

```python
with pd.ExcelWriter('SPU_optimized_orders.xlsx') as writer:  
    spuDFs[0].to_excel(writer, sheet_name='SPU - Outlet 1', index=False)
    spuDFs[1].to_excel(writer, sheet_name='SPU - Outlet 2', index=False)
    spuDFs[2].to_excel(writer, sheet_name='SPU - Outlet 3', index=False)
    writer.save()
```
